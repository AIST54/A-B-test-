```python
import pandas as pd
import numpy as np

data = pd.read_csv("item_prices.csv")
df1 = pd.read_csv("sample_a.csv")
df2 = pd.read_csv("sample_b.csv")
df3 = pd.read_csv("sample_c.csv")
data
```

```python
new_data = df3[df3.duplicated(subset=['user_id', 'item_id'])]
sample_a_2 = new_data.loc[new_data['action_id'] != 1]
sample_a_1 = new_data.loc[new_data['action_id'] == 1]
drop_row = sample_a_2.merge(sample_a_1, how='left', left_on=['user_id', 'item_id'], right_on=['user_id', 'item_id'])
drop_row = drop_row[drop_row['action_id_y'].isnull()]
drop_row = drop_row.drop('action_id_y', axis=1).rename(columns={'action_id_x':'action_id'})
df3 = pd.concat([df3, drop_row]).drop_duplicates(keep=False)
df3
```

#### Так как распределение данных не является нормальныым мы не можем применить z-test

```python
from scipy.stats import mannwhitneyu
def mannwhit_result(dfa, dfb):
    list_metric = ['ctr', 'purchase rate', 'gvm']
    results = {}
    for metric in list_metric:
        statistic, p_value = mannwhitneyu(dfa[metric], dfb[metric])
        results[metric] = {'stat':statistic, 'p_value': p_value}
    for metric, result  in results.items():
        print(f'Metric:{metric}')
        print(f'Z-statistic:{round(result["stat"], 5)}')
        print(f'P-value:{(result["p_value"])}')
        print('*' * 30)
mannwhit_result(new_df, new_df2)    
```

### Вывод: Так как нулевая гипотеза отклонена.Мы можем понять что существует разница между образцами для АА теста

```python
def mannwhit_result_new(dfa, dfb):
    list_metric = ['ctr', 'purchase rate', 'gvm']
    results = {}
    for metric in list_metric:
        statistic, p_value = mannwhitneyu(dfa[metric], dfb[metric], alternative='less')
        results[metric] = {'stat':statistic, 'p_value': p_value}
    for metric, result  in results.items():
        print(f'Metric:{metric}')
        print(f'Z-statistic:{round(result["stat"], 5)}')
        print(f'P-value:{(result["p_value"])}')
        print('*' * 30)
mannwhit_result_new(new_df, new_df1)    
```

### Вывод: По результатам AB-теста мы можем понять что есть какая то разница между сэмплами.Но то что алгоритм B улучшает работу маркетплейса мы не можем так как не все метрики показывают улучшение.Но по ключевым метрикам вроде как наблюдается прирост прибыли и есть статистическая разница 

```python
from statsmodels.stats.weightstats import ztest
def ztest_result(dfa, dfb):
    list_metric = ['ctr', 'purchase rate', 'gvm']
    results = {}
    for metric in list_metric:
        statistic, p_value = ztest(dfa[metric], dfb[metric])
        results[metric] = {'stat':statistic, 'p_value': p_value}
    for metric, result  in results.items():
        print(f'Metric:{metric}')
        print(f'Z-statistic:{round(result["stat"], 5)}')
        print(f'P-value:{(result["p_value"])}')
        print('*' * 30)
ztest_result(new_df, new_df2)   
```

```python
def ztest_result_new(dfa, dfb):
    list_metric = ['ctr', 'purchase rate', 'gvm']
    results = {}
    for metric in list_metric:
        statistic, p_value = ztest(dfa[metric], dfb[metric], alternative='smaller')
        results[metric] = {'stat':statistic, 'p_value': p_value}
    for metric, result  in results.items():
        print(f'Metric:{metric}')
        print(f'Z-statistic:{round(result["stat"], 5)}')
        print(f'P-value:{(result["p_value"])}')
        print('*' * 30)
ztest_result_new(new_df, new_df1)   
```

### Вывод: z-test показал те же результаты
